Meomery Mnagment in Python allocation in two type - stack and heap

The stack is a region of memory used for storing local variables and function call information. It operates on a Last-In-First-Out (LIFO) basis, where the most recently added item is the first to be removed.

**Heap âˆ’ Dynamic Memory Allocation**
Dynamic memory allocation occurs at runtime for objects and data structures of non-primitive types. The actual data of these objects is stored in the heap, while references to them are stored on the stack.

18. **What is the difference between os.fork() and multiprocessing in Python?**

Ans:
1. **os.fork()**
A low-level system call that creates a child process by duplicating the current process. It is available on Unix-based systems (Linux, macOS) but not on Windows.

**How it Works**
* The child process is an almost exact copy of the parent process.
* Both processes start executing from the point where the fork() call is made.
* The child process gets a unique Process ID (PID).

```python
import os

pid = os.fork()

if pid == 0:
    # Child process
    print("This is the child process.")
else:
    # Parent process
    print(f"This is the parent process. Child PID: {pid}")
```

**Key Characteristics:**
* Low-level: Requires manual management of processes.

* Platform-dependent: Works only on Unix-like systems.

* Memory Sharing: The parent and child processes do not share memory after the fork (they are isolated copies of the process).


2. **multiprocessing Module:**
A high-level module in Python's standard library for spawning and managing processes. It provides a platform-independent interface for parallelism.

* Uses the operating system's native process creation mechanisms.

* Allows creating separate processes using the Process class.

* Supports data sharing and inter-process communication via Queue, Pipe, and Value/Array objects.

```python
from multiprocessing import Process

def worker():
    print("This is a worker process.")

if __name__ == "__main__":
    p = Process(target=worker)
    p.start()  # Start the process
    p.join()   # Wait for it to finish
    print("Parent process done.")
```

Key Characteristics:

* High-level: Abstracts away the complexity of managing processes.

* Cross-platform: Works on both Unix and Windows systems.

```python
def copy_file(source_file, destination_file):
    """Copies the content of one file to another."""

    try:
        with open(source_file, 'r') as source, open(destination_file, 'w') as destination:
            for line in source:
                destination.write(line)
        print(f"File copied successfully from '{source_file}' to '{destination_file}'")
    except FileNotFoundError:
        print(f"Error: Source file '{source_file}' not found.")

if __name__ == "__main__":
    source_file = "source.txt"  # Replace with your source file name
    destination_file = "destination.txt"  # Replace with your destination file name

    copy_file(source_file, destination_file)
```

8. **Write a program to handle a file opening error using exception handling.**

```python
def read_file(file_path):
    try:
        # Attempt to open and read the file
        with open(file_path, "r") as file:
            content = file.read()
            print("File content:\n")
            print(content)
    except FileNotFoundError:
        # Handle case where the file does not exist
        print(f"Error: The file '{file_path}' does not exist.")
    except PermissionError:
        # Handle case where there are insufficient permissions to open the file
        print(f"Error: Permission denied for accessing the file '{file_path}'.")
    except Exception as e:
        # Handle any other unexpected exceptions
        print(f"An unexpected error occurred: {e}")

# Example usage
if __name__ == "__main__":
    # User-provided file path
    file_path = input("Enter the path of the file to open: ")
    read_file(file_path)
```

```python
def get_value_from_dict(dictionary, key):
    try:
        # Attempt to access the specified key
        value = dictionary[key]
        print(f"The value for key '{key}' is: {value}")
    except KeyError:
        # Handle the case where the key doesn't exist
        print(f"Error: The key '{key}' does not exist in the dictionary.")
    except Exception as e:
        # Handle any other unexpected exceptions
        print(f"An unexpected error occurred: {e}")

# Example usage
if __name__ == "__main__":
    # Sample dictionary
    sample_dict = {
        "name": "Alice",
        "age": 30,
        "city": "New York"
    }

    # Prompt user for a key to look up
    key = input("Enter the key to look up: ")
    get_value_from_dict(sample_dict, key)
```

```python
import logging

# Configure the logging system
logging.basicConfig(
    level=logging.DEBUG,  # Set the logging level to log all messages from DEBUG and above
    format="%(asctime)s - %(levelname)s - %(message)s",  # Define log message format
    filename="app.log",  # Log messages will be written to this file
    filemode="w",  # Overwrite the log file each time
)

def perform_operations():
    logging.info("Program started.")  # Log an informational message

    try:
        # Example 1: Division operation
        logging.info("Attempting to perform division.")
        numerator = 10
        denominator = 0  # This will cause a ZeroDivisionError
        result = numerator / denominator
        logging.info(f"Division successful. Result: {result}")

    except ZeroDivisionError:
        logging.error("Error: Division by zero occurred.")

    try:
        # Example 2: Dictionary key access
        logging.info("Attempting to access a dictionary key.")
        sample_dict = {"name": "Alice", "age": 30}
        value = sample_dict["city"]  # This will cause a KeyError
        logging.info(f"Key access successful. Value: {value}")

    except KeyError:
        logging.error("Error: Key 'city' does not exist in the dictionary.")

    logging.info("Program finished.")  # Log an informational message at the end

# Run the program
if __name__ == "__main__":
    perform_operations()
```

15. **Write a Python program that prints the content of a file and handles the case when the file is empty**

```python
def print_file_content(file_path):
    try:
        # Open the file in read mode
        with open(file_path, "r") as file:
            content = file.read()

            # Check if the file is empty
            if not content:
                print("The file is empty.")
            else:
                print("File content:")
                print(content)

    except FileNotFoundError:
        print(f"Error: The file '{file_path}' does not exist.")
    except PermissionError:
        print(f"Error: Permission denied for accessing the file '{file_path}'.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
if __name__ == "__main__":
    file_path = input("Enter the path of the file to read: ")
    print_file_content(file_path)
```

18. ** How would you implement a basic logging setup that logs to a file with rotation after 1MB?**

```python
import logging
from logging.handlers import RotatingFileHandler

# Create a logger
logger = logging.getLogger("MyLogger")
logger.setLevel(logging.DEBUG)  # Set the logging level to DEBUG (log everything)

# Create a rotating file handler that logs to 'my_log.log'
# Rotate the log file after it exceeds 1MB (1 * 1024 * 1024 bytes)
log_handler = RotatingFileHandler("my_log.log", maxBytes=1*1024*1024, backupCount=3)

# Create a log format
log_format = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
log_handler.setFormatter(log_format)

# Add the handler to the logger
logger.addHandler(log_handler)

# Example usage of the logger
if __name__ == "__main__":
    for i in range(10000):
        logger.debug(f"Debug message {i}")
        logger.info(f"Info message {i}")
        logger.warning(f"Warning message {i}")
        logger.error(f"Error message {i}")
        logger.critical(f"Critical message {i}")
```

21. **Write a Python program that reads a file and prints the number of occurrences of a specific word.**

```python
def count_word_occurrences(file_path, word_to_count):
    try:
        with open(file_path, 'r') as file:
            content = file.read()  # Read the entire file content
            word_count = content.lower().split().count(word_to_count.lower())  # Count the word (case-insensitive)
            return word_count
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' does not exist.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

if __name__ == "__main__":
    file_path = input("Enter the path of the file to read: ")
    word_to_count = input("Enter the word to count: ")

    word_count = count_word_occurrences(file_path, word_to_count)

    if word_count is not None:
        print(f"The word '{word_to_count}' occurs {word_count} times in the file.")
```

22. **How can you check if a file is empty before attempting to read its contents?**

```python
import os

def is_file_empty(file_path):
    # Method 1: Check file size using os.stat
    if os.stat(file_path).st_size == 0:
        return True
    return False

def read_file_if_not_empty(file_path):
    try:
        # First, check if the file is empty
        if is_file_empty(file_path):
            print("The file is empty.")
        else:
            # If the file is not empty, read and print its content
            with open(file_path, 'r') as file:
                content = file.read()
                print("File content:")
                print(content)
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' does not exist.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
if __name__ == "__main__":
    file_path = input("Enter the path of the file to read: ")
    read_file_if_not_empty(file_path)
```

23. **Write a Python program that writes to a log file when an error occurs during file handling.**

```python
import logging

# Set up logging to write to a file
logging.basicConfig(
    filename='file_handling.log',  # Log file where errors will be stored
    level=logging.ERROR,           # Only log errors and above
    format='%(asctime)s - %(levelname)s - %(message)s'  # Format of the log message
)

def read_file(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            print(content)
    except FileNotFoundError:
        logging.error(f"File '{file_path}' not found.")
        print(f"Error: The file '{file_path}' does not exist.")
    except PermissionError:
        logging.error(f"Permission denied when trying to read '{file_path}'.")
        print(f"Error: Permission denied when trying to read '{file_path}'.")
    except Exception as e:
        logging.error(f"An unexpected error occurred while reading '{file_path}': {e}")
        print(f"An unexpected error occurred: {e}")

def write_file(file_path, content):
    try:
        with open(file_path, 'w') as file:
            file.write(content)
            print(f"Content successfully written to {file_path}")
    except PermissionError:
        logging.error(f"Permission denied when trying to write to '{file_path}'.")
        print(f"Error: Permission denied when trying to write to '{file_path}'.")
    except Exception as e:
        logging.error(f"An unexpected error occurred while writing to '{file_path}': {e}")
        print(f"An unexpected error occurred: {e}")

# Example usage
if __name__ == "__main__":
    # Attempt to read a non-existent file to log an error
    file_to_read = "non_existent_file.txt"
    read_file(file_to_read)

    # Attempt to write to a file
    file_to_write = "output.txt"
    content_to_write = "This is some sample content"
    write_file(file_to_write, content_to_write)
```
